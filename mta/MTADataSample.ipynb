```python
calendar_weekday = dfs["calendar"][dfs["calendar"].friday + dfs["calendar"].monday == 2].drop_duplicates()
```

```python
trips_weekday = dfs["trips"].merge(right = calendar_weekday[["service_id"]], on = ["service_id"], how = "inner")
```

```python
trips_weekday_w_route_color = trips_weekday.merge(
    right = dfs["routes"][["route_id", "route_color"]].drop_duplicates(),
    on = ["route_id"],
    how = "left")
```

```python
trips_weekday_w_route_color.head()
```

```python
stop_times_weekday = dfs["stop_times"][["trip_id", "arrival_time", "departure_time", "stop_id", "stop_sequence"]].merge(
    right = trips_weekday_w_route_color[["trip_id", "route_id", "route_color"]].drop_duplicates(),
    on = ["trip_id"],
    how = "right").merge(
    right = dfs["stops"][["stop_id", "stop_lat", "stop_lon"]].drop_duplicates(),
    on = ["stop_id"],
    how = "left")
```

```python
import numpy as np
```

```python
sum(np.prod(i) for i in zip([3600, 60, 1], [5, 59, 59]))
```

```python
def sec_since_midnight(tim_str):
    return sum(np.prod(i) for i in zip([3600, 60, 1], [int(i) for i in tim_str.split(":")]))

stop_times_weekday["arrive"] = stop_times_weekday.arrival_time.map(sec_since_midnight)
stop_times_weekday["depart"] = stop_times_weekday.departure_time.map(sec_since_midnight)
```

```python
next_cols = ["arrive", "depart", "stop_id", "stop_lat", "stop_lon"]
```

```python
for col in next_cols:
    stop_times_weekday["next_{}".format(col)] = stop_times_weekday.sort_values("stop_sequence").groupby(["trip_id"])[col].shift(-1)
```
