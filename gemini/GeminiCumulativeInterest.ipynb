```python
import pandas as pd
```

```python
events = pd.read_csv("./gemini_rel_events.csv")
```

```python
inside = pd.read_json("./gemini_inside.json")
```

```python
events.head()
```

```python
events["signed_price_padded"] = events.side.map(lambda x: -1 if x == "bid" else 1) * events.price_padded
```

```python
events["cum_interest"] = events.sort_values("signed_price_padded").groupby(["ref_tim", "side"], as_index = False)["remaining"].cumsum()
```

```python
events["price_padded_str"] = events.price_padded.astype("str").str.pad(7, "right", "0")
```
