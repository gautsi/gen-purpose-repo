```python
ad = requests.get("https://web.enrboenyc.us/CD22509AD0.html")
ad_soup = BeautifulSoup(ad.text)
```

```python
ad_soup
```

```python
res_table = ad_soup.findAll("table", attrs={"class": "underline"})
```

```python
len(res_table)
```

```python
res_table0 = res_table[0]
```

```python
rows = res_table0.findAll("tr")
```

```python
len(rows)
```

```python
rows[0]
```

```python
len(rows[0].findAll("td"))
```

```python
rows_cols = [[i.text for i in row.findAll("td")] for row in rows]
```

```python
set([len(i) for i in rows_cols])
```

```python
rows_cols
```

```python
all_candidates = [
    {
        "candidate": t,
        "col_num": i,
        "party": rows_cols[1][i],
        "votes": {}
    } for i, t in enumerate(rows_cols[0]) if t != "\xa0"]
```

```python
all_candidates
```

```python
for r in rows_cols:
    if "AD" in r[0]:
        for cand in all_candidates:
            cand["votes"][re.findall("\d{2}", r[0])[0]] = r[cand["col_num"]]
```

```python
all_candidates
```

```python
with open("./nyc_pub_adv.json", "w") as output:
    json.dump(all_candidates, output)
```
