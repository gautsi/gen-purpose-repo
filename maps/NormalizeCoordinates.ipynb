```python
import pandas as pd
```

```python
unt = pd.read_csv("./untangled_county_squares.csv")
```

```python
def get_min_max(df, center_col_name = "prop_scaled_{}_rank", side_col_name = "sqrt_mult"):
    for si, i in enumerate(["min", "max"]):
        for j in ["lat", "lng"]:
            df[("{}_" + center_col_name).format(i, j)] = (
                df[center_col_name.format(j)] + (2 * si - 1) *
                (df[side_col_name] / 2))
    return df
```

```python
check_cols = [
    "county",
    "lat_rank",
    "lng_rank",
    "sqrt_mult",
    "min_new_scaled_lat_rank",
    "new_scaled_lat_rank",
    "max_new_scaled_lat_rank",
    "min_new_scaled_lng_rank",
    "new_scaled_lng_rank",
    "max_new_scaled_lng_rank"
]
unt[unt.county == "Kings"][check_cols]
```

```python
def normalize(df, col_names, min_col_name, max_col_name):
    for c in col_names:
        df["norm_" + c] = (df[c] - df[min_col_name].min()) / (df[max_col_name].max() - df[min_col_name].min())
    return df
```

```python
norm_cols = ([{
    "col_names": ["{}_rank".format(i)],
    "min_col_name": "{}_rank".format(i),
    "max_col_name": "{}_rank".format(i)} for i in ["lat", "lng"]] + 
    [{
    "col_names": ["{}new_scaled_{}_rank".format(j, i) for j in ["", "min_", "max_"]],
    "min_col_name": "min_new_scaled_{}_rank".format(i),
    "max_col_name": "max_new_scaled_{}_rank".format(i)} for i in ["lat", "lng"]])

for i in norm_cols:
    unt = normalize(unt, col_names = i["col_names"], min_col_name = i["min_col_name"], max_col_name = i["max_col_name"])
```

```python
unt.describe()
```

```python
counties = unt[["county"]].drop_duplicates().reset_index()
counties["county_id"] = counties.index
```

```python
unt_w_county_id = unt.merge(right = counties[["county", "county_id"]], on = ["county"], how = "left")
```

```python
norm_col_names = ["norm_" + i for j in [i["col_names"] for i in norm_cols] for i in j]
```

```python
unt_w_county_id[["square_id", "county", "county_id"] + norm_col_names].to_json(
    "./untangled_norm_county_squares.json", orient = "records")
```
