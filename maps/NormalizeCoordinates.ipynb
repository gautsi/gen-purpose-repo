```python
import pandas as pd
```

```python
unt = pd.read_csv("./untangled_county_squares.csv")
```

```python
def normalize(df, col_names):
    for c in col_names:
        df["norm_" + c] = df[c] / df[c].max()
    return df
```

```python
norm_cols = ["sqrt_mult"] + ["{}_rank".format(i) for i in ["lat", "lng"]] + ["new_scaled_{}_rank".format(i) for i in ["lat", "lng"]]
```

```python
unt = normalize(unt, norm_cols)
```

```python
unt.describe()
```

```python
counties = unt[["county"]].drop_duplicates().reset_index()
```

```python
counties["county_id"] = counties.index
counties
```

```python
unt[["square_id", "county"] + ["norm_" + i for i in norm_cols]].to_json("./untangled_norm_county_squares.json", orient = "records")
```
