```python
import pandas as pd
import json
import matplotlib.pyplot as plt
```

```python
import requests
```

```python
%matplotlib inline
```

```python
a = requests.get("https://raw.githubusercontent.com/gautsi/gautsi.github.io/master/assets/updated_marks.json")
```

```python
ny_counties = json.loads(a.text)
```

```python
len(ny_counties)
```

```python
ny_counties[0]
```

```python
from matplotlib import cm
```

```python
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
fig, ax = plt.subplots(figsize = (15, 15))
s = ax.scatter(
    x = [i["center"][0] for i in ny_counties],
    y = [-1 * i["center"][1] for i in ny_counties],
    marker = "s",
    c = [i["totalVotes"] for i in ny_counties],
    cmap = cm.viridis,
    s = 3700,
    alpha = 0.4,
    edgecolor = "none")

for c in ny_counties:
    ax.annotate(
        c["name"],
        xy = (c["pos"][0], -1 * c["pos"][1]),
        xytext = (c["pos"][0] + 1, -1 * c["pos"][1] - 35),
        color = "steelblue",
        fontsize = 11,
        fontweight = "bold")
    

divider = make_axes_locatable(ax)
cax = divider.append_axes('right', size='1%', pad=0.05)
fig.colorbar(s, cax = cax)
```
